```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
z = np.cos(np.linspace(0, 2*np.pi, 7, endpoint= False))
```

```python
z
```

```python
plt.stem(x, z, use_line_collection = True)
plt.show()
```

```python
z.sort()
plt.stem(x, z, markerfmt = 's', linefmt='--', basefmt = ':', use_line_collection=False)
plt.show()
```

```python
plt.stem(x, z, markerfmt = 's', linefmt='--', basefmt = ':', use_line_collection=True)
plt.show()
```

```python
plt.hlines(y=x, xmin = 0 , xmax = z, color='tomato')
plt.plot(z, x, "o")
plt.show()
```

```python
f, axs = plt.subplots(1,2,figsize=(8,4))
axs[0].barh(x,z)
axs[1].hlines(y=x, xmin = 0 , xmax = z, color='skyblue')
axs[1].plot(z, x, marker = "s", linewidth = 0.5)
plt.show()
```
