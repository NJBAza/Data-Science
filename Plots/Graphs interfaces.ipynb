```python
import numpy as np
import matplotlib

matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt

%matplotlib inline

x = np.arange(0,np.pi/2,0.1)   # start,stop,step
y = np.arcsin(0.75*np.sin(x))

fig, ax = plt.subplots(figsize=(5, 4), tight_layout=True)
ax.plot(x, y)

ax.set_xlabel(r'Angle of incidence for $\theta\in\left[0,\frac{\pi}{2}\right]$', color='r', fontsize=16)
ax.set_ylabel(r'Angle of refraction', color='g', fontsize=16)
ax.set_title(r'Graph for air–water interface case', fontsize=20, color='brown')

plt.show()
```
